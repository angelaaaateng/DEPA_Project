```python
import glob
import pandas  as pd

path1 = "/Users/angelateng/Documents/GitHub/DEPA_Project/Tweets_Labeled"

all_crypto = glob.glob(path1 + "/*.csv")

li = []

for filename in all_crypto:
    labeled_crypto = pd.read_csv(filename, index_col=None, header=0)
    li.append(labeled_crypto)

crypto_labeled_data = pd.concat(li, axis=0, ignore_index=True)

print(crypto_labeled_data.shape)
```

```python

path2 = "/Users/angelateng/Documents/GitHub/DEPA_Project/Tweets_All"

all_crypto_no_labels = glob.glob(path2 + "/*.csv")

li = []

for filename in all_crypto_no_labels:
    unlabeled_crypto = pd.read_csv(filename, index_col=None, header=0)
    li.append(unlabeled_crypto)

crypto_unlabeled_data = pd.concat(li, axis=0, ignore_index=True)

print(crypto_unlabeled_data.shape)
```
